```python
#9.2.3
query = """
SELECT
  m.id AS match_id,
  c.name AS country_name,
  l.name AS league_name,
  m.season,
  m.stage,
  m.date,
  ht.team_long_name AS home_team,
  at.team_long_name AS away_team,
  m.home_team_goal,
  m.away_team_goal
FROM Match m
JOIN Country c ON m.country_id = c.id
JOIN League l ON m.league_id = l.id
JOIN Team ht ON m.home_team_api_id = ht.team_api_id
JOIN Team at ON m.away_team_api_id = at.team_api_id
WHERE c.name = 'Spain'
ORDER BY m.date ASC
LIMIT 10;
"""

detailed_matches = pd.read_sql(query, conn)
detailed_matches
```

```python
#10.1.2
query = """
SELECT
  c.name AS Country,
  l.name AS League,
  COUNT(DISTINCT m.stage) AS Stages,
  m.season AS Season,
  COUNT(DISTINCT ht.team_api_id) + COUNT(DISTINCT at.team_api_id) AS number_of_teams,
  ROUND(AVG(m.home_team_goal), 2) AS avg_local_goals,
  ROUND(AVG(m.away_team_goal), 2) AS avg_visitor_goals,
  ROUND((SUM(m.home_team_goal + m.away_team_goal) / COUNT(m.id)), 2) AS avg_goals,
  ROUND(AVG((m.home_team_goal - m.away_team_goal)), 2) AS avg_goal_dif,
  SUM(m.home_team_goal + m.away_team_goal) AS total_goals
FROM Match m
JOIN Country c ON m.country_id = c.id
JOIN League l ON m.league_id = l.id
JOIN Team ht ON m.home_team_api_id = ht.team_api_id
JOIN Team at ON m.away_team_api_id = at.team_api_id
WHERE c.name IN ('Spain', 'Germany', 'France', 'Italy', 'England')
GROUP BY c.name, l.name, m.season
HAVING ROUND((SUM(m.home_team_goal + m.away_team_goal) / COUNT(m.id)), 2) > 2.5;
"""

leages_by_season = pd.read_sql(query, conn)
leages_by_season
```

```python
#10.1.3
query = """
SELECT
  c.name AS Country,
  l.name AS League,
  ht.team_long_name AS Team,
  COUNT(m.stage) AS Stages,
  m.season AS Season,
  SUM(m.home_team_goal) AS sum_goals,
  ROUND(AVG(m.home_team_goal), 2) AS avg_goals
FROM Match m
JOIN Country c ON m.country_id = c.id
JOIN League l ON m.league_id = l.id
JOIN Team ht ON m.home_team_api_id = ht.team_api_id
WHERE c.name IN ('Spain', 'Germany', 'England')
GROUP BY c.name, l.name, ht.team_long_name, m.season
HAVING SUM(m.home_team_goal) > 50
ORDER BY m.season DESC, sum_goals DESC;
"""

teams_by_season = pd.read_sql(query, conn)
teams_by_season
```

```python
#12.2
query = """
SELECT
    M.date AS MatchDate,
    T1.team_long_name AS HomeTeam,
    T2.team_long_name AS VisitorTeam,
    CASE
        WHEN M.home_team_api_id = 8634 AND M.home_team_goal > M.away_team_goal THEN 'FC Barcelona gana'
        WHEN M.home_team_api_id = 8633 AND M.away_team_goal > M.home_team_goal THEN 'Real Madrid gana'
        ELSE 'Empate'
    END AS Outcome
FROM Match M
JOIN Team T1 ON M.home_team_api_id = T1.team_api_id
JOIN Team T2 ON M.away_team_api_id = T2.team_api_id
WHERE (T1.team_api_id = 8634 AND T2.team_api_id = 8633);
"""

BarcaMadrid_results = pd.read_sql(query, conn)
BarcaMadrid_results
```

```python
#13.2
query = """
SELECT
    TA.date AS RatingDate,
    TA.team_api_id AS TeamID,
    T.team_long_name AS Team,
    CASE
        WHEN TA.buildUpPlaySpeed >= 70 THEN 'Fast'
        WHEN TA.buildUpPlaySpeed <= 30 THEN 'Slow'
        ELSE 'Balanced'
    END AS PlaySpeed,
    CASE
        WHEN TA.buildUpPlayPassing >= 70 THEN 'Long'
        WHEN TA.buildUpPlayPassing <= 30 THEN 'Short'
        ELSE 'Mixed'
    END AS PlayPassing,
    CASE
        WHEN TA.chanceCreationShooting >= 68 THEN 'Lots'
        WHEN TA.chanceCreationShooting <= 30 THEN 'Little'
        ELSE 'Normal'
    END AS CreationShooting
FROM Team_Attributes TA
JOIN Team T ON TA.team_api_id = T.team_api_id
GROUP BY TA.date, TA.team_api_id, T.team_long_name
ORDER BY TA.date, T.team_long_name;
"""

subquery = pd.read_sql(query, conn)
subquery
```

```python
#13.3
query = """
SELECT
    TA.date AS RatingDate,
    TA.team_api_id AS TeamID,
    T.team_long_name AS Team,
    CASE
        WHEN TA.buildUpPlaySpeed >= 70 THEN 'Fast'
        WHEN TA.buildUpPlaySpeed <= 30 THEN 'Slow'
        ELSE 'Balanced'
    END AS PlaySpeed,
    CASE
        WHEN TA.buildUpPlayPassing >= 70 THEN 'Long'
        WHEN TA.buildUpPlayPassing <= 30 THEN 'Short'
        ELSE 'Mixed'
    END AS PlayPassing,
    CASE
        WHEN TA.chanceCreationShooting >= 68 THEN 'Lots'
        WHEN TA.chanceCreationShooting <= 30 THEN 'Little'
        ELSE 'Normal'
    END AS CreationShooting
FROM Team_Attributes TA
JOIN Team T ON TA.team_api_id = T.team_api_id
WHERE
    TA.buildUpPlaySpeed IS NOT NULL AND
    TA.buildUpPlayPassing IS NOT NULL AND
    TA.chanceCreationShooting IS NOT NULL
GROUP BY TA.date, TA.team_api_id, T.team_long_name
ORDER BY TA.date, T.team_long_name;
"""

subquery = pd.read_sql(query, conn)
subquery
```
